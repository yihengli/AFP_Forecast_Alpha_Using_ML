```python
import pandas as pd
import numpy as np
import os
```

```python
path = "/Users/valentinlepinay/dev/AFP_Forecast_Alpha_Using_ML/stock/"
paths = [path + x for x in os.listdir(path) if 'DS_Store' not in x]
```

```python
# _is = False
for path in paths:
    print(path)
    if path == "/Users/valentinlepinay/dev/AFP_Forecast_Alpha_Using_ML/stock/nan.csv":
#         _is = True
        continue
#     if not _is:
#         continue
        
    df = pd.read_csv(path, usecols=['PERMNO', 'date', 'TICKER', 'sprtrn', 'RET'],
                     dtype={'PERMNO': np.str, 'date': np.str, 'TICKER': np.str})
    df.date = pd.to_datetime(df.date, format="%Y/%m/%d")
    df.sort_values('date', inplace=True)
    df.reset_index(drop=True, inplace=True)
    df.RET = pd.to_numeric(df.RET, errors='coerce', downcast='float')
    df.sprtrn = pd.to_numeric(df.sprtrn, errors='coerce', downcast='float')
    df.rename(columns={'RET': 'C2C', 'sprtrn': 'indexC2C'}, inplace=True)
    df['C2CindexAdj'] = df.C2C.values - df.indexC2C.values
    df['C2CindexAdjFwd3Days'] = (df.C2C + 1).rolling(3).agg(lambda x : x.prod()) - \
        (df.indexC2C + 1).rolling(3).agg(lambda x : x.prod())
    df['C2CindexAdjFwd5Days'] = (df.C2C + 1).rolling(5).agg(lambda x : x.prod()) - \
        (df.indexC2C + 1).rolling(5).agg(lambda x : x.prod())
    df['C2CindexAdjFwd10Days'] = (df.C2C + 1).rolling(10).agg(lambda x : x.prod()) - \
        (df.indexC2C + 1).rolling(10).agg(lambda x : x.prod())

    df['C2CindexAdjFwd1Day'] = df.C2CindexAdj.shift(-1)
    df['C2CindexAdjFwd3Days'] = df.C2CindexAdjFwd3Days.shift(-1)
    df['C2CindexAdjFwd5Days'] = df.C2CindexAdjFwd5Days.shift(-1)
    df['C2CindexAdjFwd10Days'] = df.C2CindexAdjFwd10Days.shift(-1)
    df.date = df.date.dt.strftime('%Y%m%d')
#     df.to_csv(path.replace('stock', 'C2C'), index=False, float_format='%.8f')
```
